```python
import json
import pandas as pd
from pathlib import Path
import glob
```

```python
results_dir = Path("../reports/results")
```

```python
result_files = glob.glob(str(results_dir / "*.json"))
```

```python
result_files
```

```python
result_list = []
for result_file in result_files:
    with open(result_file, "r") as f:
        data = json.load(f)

    result_list.append(data)

    # Save the DataFrame to a CSV file
    
df = pd.DataFrame(result_list)
```

```python
df
```

```python
df.to_csv(results_dir / "results.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sorted_df = df.sort_values(by='avg_test_acc', ascending=False)
```

```python
sorted_df
```

```python
std_df = df.sort_values(by='std_test_acc', ascending=True)
```

```python
std_df
```

```python
sns.set_theme(style="whitegrid")


def plot_results(df):
    plt.figure(figsize=(10, 6))
    df_sorted = df.sort_values(by='model_name')
    ordered_names = df_sorted['model_name'].tolist()
    sns.barplot(x="model_name", y="avg_test_acc", data=df, hue="family", order=ordered_names)
    plt.title("Model Accuracy Comparison")
    plt.xlabel("Model")
    plt.ylabel("Accuracy")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(results_dir / "model_accuracy_comparison.png")
    plt.show()
```

```python
plot_results(df)
```

```python
def plot_scatter_results(df):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x="avg_test_acc", y="std_test_acc", data=df, hue="family", size="model_params", sizes=(20, 200), alpha=0.7)
    plt.title("Model Accuracy vs Std Dev Comparison")
    plt.xlabel("Std Dev of Accuracy")
    plt.ylabel("Accuracy")
    plt.tight_layout()
    plt.savefig(results_dir / "model_accuracy_vs_std_comparison.png")
    plt.show()
```

```python
plot_scatter_results(df)
```

```python
df_zoomed_in = df[df['family'] != "inception"]
```

```python
plot_scatter_results(df_zoomed_in)
```
